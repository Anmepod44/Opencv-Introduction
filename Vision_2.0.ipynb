```python
 def findCorners(img):
        #finds corners in an image using the cv2.goodFeaturesToTrack function of opencv
        
        if(img.shape[2]==3):
            gray=cv2.cvtColor(img.copy(),cv2.COLOR_BGR2GRAY)
        
        corners=cv2.goodFeaturesToTrack(gray,1000,0.01,1)
        corners=np.int0(corners)
            
      
            
        for i in corners:
            x,y = i.ravel()
            cv2.circle(img,(x,y),3,(255,34,0),1)
                
        return img
                
```

```python
def draw_bounding_box(img):
    
    #Perform a Threshold of the image convert it to gray scale prior to processing
    
    
    frame=img
    
    
    img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
#     img=cv2.Canny(img,100,200);
    
    #Thresholding paramters are calculated for different values and regions of the image.
    img=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
    
    
    contours,hierarchy=cv2.findContours(img,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    for i in contours:
        
        if(cv2.contourArea(i)>100):
            
            #Draw a bounding box around the object.
            x,y,w,h=cv2.boundingRect(i)
            frame=cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),1)
            
            
            #draw a rotating rectangle around the object 
            rect=cv2.minAreaRect(i)
            box=cv2.boxPoints(rect)
            box=np.int0(box)
            frame=cv2.drawContours(frame,[box],0,(0,0,255),1)
            
            
            # draw a circle  on the image:
            center,radius=cv2.minEnclosingCircle(i)
            center=(int(center[0]),int(center[1]))
            radius=int(radius)
            frame=cv2.circle(frame,center,radius,(255,0,0),1)
            
            
    
    
    return (frame,img)
            
    
    
    
    
```

```python
 #writing a class to display the trackbar window that will be used throughout the application.
class Trackbar(object):
 
    def __init__(self):
        
        self.width=640
        self.height=240
        self.cap=cv2.VideoCapture(0)

    #width
    def setwidth(self,_):
        self.cap.set(1,_)
        
    #height
    def setheight(self,_):
        self.cap.set(2,_)
     
    def display(self,img):
        
         if cv2.waitKey(1)!=ord('q'):
                cv2.imshow("This is the frame",img)
        
    def run(self):
        while True:
            _,frame=cap.read()
            self.display(frame)
            cv2.namedWindow("MyWindow")
            cv2.resizeWindow("MyWindow",640,240)
            cv2.createTrackbar("Width","MyWindow",0,179,self.setheight)
            cv2.createTrackbar("Height","MyWindow",255,255,self.setwidth)
            
            
           
            if cv2.waitKey()==ord('q'):
                break
        cv2.destroyAllWindows()
```
